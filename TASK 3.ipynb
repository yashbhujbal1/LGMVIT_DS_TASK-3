## Data Preprocessing

```python
X=df.drop(columns=['Species'])
y=df['Species']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

## Splitting the dataset into training and testing set

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

## Building the Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(criterion='entropy')
```

```python
dt=dt.fit(X_train,y_train)
```

```python
score=accuracy_score(y_test,dt.predict(X_test))
print('Accuracy of the model is: ',(score*100))
```

```python
from sklearn.tree import export_graphviz
from io import StringIO
from IPython.display import Image
import pydotplus
```

```python
data=StringIO()
export_graphviz(dt,out_file=data,filled=True,rounded=True,special_characters=True,feature_names=X.columns.values.tolist(),class_names='Species')
graph=pydotplus.graph_from_dot_data(data.getvalue())
graph.write_png('Iris.png')
Image(graph.create_png())
```
